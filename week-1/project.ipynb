```python
# Program to calculate Simple Interest
P = float(input("Enter principal amount: "))
R = float(input("Enter annual interest rate (in %): "))
T = float(input("Enter time in years: "))

# Formula: A = P * (1 + (R/100) * T)
A = P * (1 + (R / 100) * T)

print(f"The total amount after {T} years is: {A:.2f}")
```

```python
# Program to calculate Compound Interest
P = float(input("Enter principal amount: "))
R = float(input("Enter annual interest rate (as decimal, e.g., 0.05 for 5%): "))
n = int(input("Number of times interest compounded per year: "))
t = float(input("Number of years: "))

# Formula: A = P * (1 + R/n)**(n*t)
A = P * (1 + R / n)**(n * t)

print(f"The compound interest amount is: {A:.2f}")
```

```python
# Program to calculate Future Value of an Annuity
PMT = float(input("Enter periodic payment: "))
R = float(input("Enter annual interest rate (as decimal): "))
n = int(input("Number of payments per year: "))
t = float(input("Number of years: "))

# Formula: A = PMT * (((1 + R/n)**(n*t) - 1) / (R/n))
rate_per_period = R / n
periods = n * t
A = PMT * (((1 + rate_per_period)**periods - 1) / rate_per_period)

print(f"The total value of the annuity is: {A:.2f}")
```
