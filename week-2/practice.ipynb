```python
str = input("Enter your state of origin:")

print ("Your state of origin is: ", str)
print ("The first character is: ", str[0])
print ("The characters starting from 3rd to 5th are: ", str[2:5])
print ("The string starting from 3rd character is: ", str[2:])
print ("State of origin two times", str * 2)
```

```python
# Input from user
m = float(input("Enter mass in kilograms: "))

# Constant value for the speed of light in m/s
c = 299792458

# Calculating energy using Einstein's equation
energy = m * c ** 2

# Displaying the result
print(f"The energy equivalent to {m} kg of mass is {energy} joules.")
```

```python
my_list = ['Anaconda', 786, 2.23, 'Jupyter', 70.2]
shortlist = [321, 'Python']

# Prints complete list
print(my_list)

# Prints first element of the list
print(my_list[0])

# Prints elements starting from 2nd till 3rd
print(my_list[1:3])

# Prints elements starting from 3rd element onwards
print(my_list[2:])

# Prints shortlist two times
print(shortlist * 2)

# Prints concatenated lists
print(my_list + shortlist)
```

```python
tuple = ("Ekiti", 750, 'Oshogbo', 250, "Akure", 500)
s_tuple = ("Abeokuta", 300, "Ogbomoso")

# Prints the complete tuple
print (tuple)

# Prints last element of the tuple
print (tuple[-1])

# Prints elements of the tuple starting from 2nd till 3rd
# Note: index 2 is the 3rd item, index 4 is the 5th (excluded)
print (tuple[2:4])

# Prints elements of the tuple starting from 3rd element
print (tuple[3:])

# Prints the contents of the tuple three times
print (s_tuple * 3)

# Prints concatenated tuples
print (tuple + s_tuple)
```

```python
# Screenshot: 191959.png (Booleans)
game_1, game_2 = 2, 4
print(bool(game_1 == game_2)) # False

val = None
print(bool(val)) # False

num_tuple = ()
print(bool(num_tuple)) # False (empty sequence)

age = True
print(bool(age)) # True

# Screenshot: 192012.png & 192021.png (Casting)
# Int Casting
grade_int = int(70)
gpa_int = int(4.9) # Note: This will be 4, as int() truncates
cgpa_int = int("4")
print(f"Grade = {grade_int}, GPA = {gpa_int}, CGPA = {cgpa_int}")

# Float Casting
grade_flt = float(97)
gpa_flt = float(5)
cgpa_flt = float("4.7")
print(f"Grade = {grade_flt}, GPA = {gpa_flt}, CGPA = {cgpa_flt}")
```

```python
# Convert to int

grade = int(70)      # grade will be 70
gpa = int(4.9)        # gpa will be 4.9
cgpa = int("4")       # cgpa will be 4

print(f"Grade = {grade}")
print(f"GPA = {gpa}")
print(f"CGPA = {cgpa}")
```
