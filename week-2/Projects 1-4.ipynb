```python
def swap_ages():
    # Input two names
    name1 = input("Enter first name: ")
    age1 = int(input(f"Enter age for {name1}: "))
    
    name2 = input("Enter second name: ")
    age2 = int(input(f"Enter age for {name2}: "))

    print(f"\nBefore Swap: {name1} is {age1}, {name2} is {age2}")

    # Swapping logic
    temp = age1
    age1 = age2
    age2 = temp

    print(f"After Swap: {name1} is {age1}, {name2} is {age2}")

swap_ages()
```

```python
import numpy as np

def find_roots():
    print("--- Polynomial Root Finder ---")
    print("1. Quadratic Equation (Ax^2 + Bx + C = 0)")
    print("2. Cubic Equation (Ax^3 + Bx^2 + Cx + D = 0)")
    print("3. Quartic Equation (Ax^4 + Bx^3 + Cx^2 + Dx + E = 0)")
    
    choice = input("\nSelect the operation (1, 2, or 3): ")

    if choice == '1':
        a = float(input("Enter A: "))
        b = float(input("Enter B: "))
        c = float(input("Enter C: "))
        roots = np.roots([a, b, c])
    elif choice == '2':
        a = float(input("Enter A: "))
        b = float(input("Enter B: "))
        c = float(input("Enter C: "))
        d = float(input("Enter D: "))
        roots = np.roots([a, b, c, d])
    elif choice == '3':
        a = float(input("Enter A: "))
        b = float(input("Enter B: "))
        c = float(input("Enter C: "))
        d = float(input("Enter D: "))
        e = float(input("Enter E: "))
        roots = np.roots([a, b, c, d, e])
    else:
        print("Invalid selection.")
        return

    print(f"\nThe roots are: {roots}")

find_roots()
```
