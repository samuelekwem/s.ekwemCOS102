```python
import tkinter as tk
from tkinter import messagebox
import pandas as pd

# Function to verify employee and show department members
def verify_employee():
    user_name = name_entry.get().strip()
    user_dept = dept_entry.get().strip()

    try:
        # Load the dataset
        df = pd.read_csv('GIG-logistics.csv')
        
        # Clean column names and data to ensure matching works
        df.columns = df.columns.str.strip()
        
        # Check if employee exists in the specific department
        # Assumes CSV has 'NAME' and 'DEPARTMENT' columns
        employee_match = df[(df['NAME'].str.lower() == user_name.lower()) & 
                            (df['DEPARTMENT'].str.lower() == user_dept.lower())]

        if not employee_match.empty:
            # Welcome the employee
            welcome_text = f"Welcome, {user_name}!\n\nOther members in {user_dept}:"
            
            # Get other members of the same department
            dept_members = df[df['DEPARTMENT'].str.lower() == user_dept.lower()]['NAME'].tolist()
            dept_members.remove(employee_match['NAME'].values[0]) # Remove the current user from the list
            
            # Create a new window to display results
            result_window = tk.Toplevel(root)
            result_window.title("Success")
            result_window.geometry("400x300")
            
            tk.Label(result_window, text=welcome_text, font=("Arial", 10, "bold")).pack(pady=10)
            
            # Display list of colleagues
            listbox = tk.Listbox(result_window, width=40)
            listbox.pack(padx=20, pady=10)
            for member in dept_members:
                listbox.insert(tk.END, member)
        else:
            messagebox.showerror("Error", "Employee does not exist or Department is incorrect.")
            
    except FileNotFoundError:
        messagebox.showerror("System Error", "Dataset 'GIG-logistics.csv' not found!")

# Main Window Setup
root = tk.Tk()
root.title("GIG Logistics Name Verification")
root.geometry("400x250")

# UI Elements
tk.Label(root, text="Employee Name:").pack(pady=(20, 0))
name_entry = tk.Entry(root, width=30)
name_entry.pack(pady=5)

tk.Label(root, text="Department:").pack(pady=(10, 0))
dept_entry = tk.Entry(root, width=30)
dept_entry.pack(pady=5)

verify_button = tk.Button(root, text="Verify Identity", command=verify_employee, 
                          bg="blue", fg="white", font=("Arial", 9, "bold"))
verify_button.pack(pady=20)

root.mainloop()
```
